```python
def get_data(symbol="MSFT", filename="split"):
	df = pd.DataFrame()
	try:
		df = pd.read_csv(f'data/{filename}/{symbol}.csv')
	except Exception as e:
		qt.log.warn(f"{filename} file not found for {symbol}")
	qt.log.info(f"df.shape = {df.shape}")
	return df
```

```python
ts = get_data(symbol=symbol, filename='split')
td = get_data(symbol=symbol, filename='div')
ti = get_data(symbol=symbol, filename='intraday')
```

```python
import requests
from io import StringIO

def download_file_by_date(query_date=dt.date(2025,1,1), filename="dividends", api_token=EODHD_API_KEY):
	df = pd.DataFrame()
	try:
		date_str = query_date.strftime("%Y-%m-%d")
		url = f'https://eodhd.com/api/eod-bulk-last-day/US?api_token={api_token}&date={date_str}&type={filename}&fmt=csv'
		data = requests.get(url)
		df = pd.read_csv(StringIO(data.content.decode()))

	except Exception as e:
		qt.log.warning(f"couldnt load {filename} data for date {query_date}")

	return df
```

```python
td = download_file_by_date(dt.date(2023, 12, 27), filename="dividends")
dates = pd.date_range(start_dt, end_dt).date
for d in dates:
	td = download_file_by_date(query_date=d, )
```

```python
intraday_df = ti.copy()
splits_df = ts.copy()
div_df = td.copy()
```

```python
df = intraday_df.copy()
df["datetime"] = pd.to_datetime(df["timestamp"], utc=True, unit="s")
df["datetime_us"] = df["datetime"].dt.tz_convert("America/New_York")
df["date"] = df["datetime_us"].dt.date
df["time"] = df["datetime_us"].dt.time
```

```python
# keep the last close (filter between 09:30 and 16:00 pm)
df_close = df[(df['time'] >= dt.time(9,30)) & (df['time'] <= dt.time(16,0))].sort_values(['date', 'time']).reset_index(drop=True)
df_close = pd.DataFrame(df_close.groupby('date')['close'].last()).reset_index()

df_close.set_index('date')['close'].plot()
```

```python
start_date = df_close['date'].min()
end_date = df_close['date'].max()
tmp = pd.DataFrame(pd.date_range(start_date, end_date).date, columns=['date'])
df_close = pd.merge(tmp, df_close, on='date', how='left')

# split
splits = splits_df.copy()
splits["date"] = pd.to_datetime(splits["date"]).dt.date
splits["date"] = splits["date"].apply(lambda x: x-dt.timedelta(1))
splits["ratio"] = splits["split"].apply(lambda x: float(x.split("/")[0]) / float(x.split("/")[1]))

# add split
df_close = pd.merge(df_close, splits[['date', 'ratio']], on='date', how='left')
df_close['ratio'] = df_close['ratio'].fillna(1)
df_close = df_close.sort_values('date', ascending=False).reset_index(drop=True)
df_close['split_adj'] = 1/df_close['ratio'].cumprod()


df_close['close_adj'] = df_close['close']*df_close['split_adj']
```
